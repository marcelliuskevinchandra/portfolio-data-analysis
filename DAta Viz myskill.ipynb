```python
import numpy as np #library untuk matematika
import pandas as pd #library untuk dataframe
import seaborn as sns #library untuk visualisasi
import matplotlib.pyplot as plt
```

Load data

```python
dataset = sns.load_dataset("mpg")
```

```python
dataset.head()
```

Mempersiapkan data untuk plot

```python
data_origin = dataset["origin"].value_counts().to_dict()

data_pie = data_origin.values()
label_pie = data_origin.keys()
colors = sns.color_palette('pastel')[0:5]
```

Pie chart

```python
plt.pie(x = data_pie,
        labels = label_pie,
        colors = colors,
        autopct = '%.0f%%'
        )
plt.title("Origin")
plt.show()
```

Membuat histogram

```python
sns.histplot(data = dataset,
             x = "mpg",
             bins = 30)
plt.show()
```

Membuat boxplot

```python
sns.boxplot(data = dataset,
            x = "mpg")
plt.show()
```

Membuat bar chart

```python
sns.barplot(data = dataset,
            x = "origin",
            y = "mpg",
            )
plt.show()
```

Membuat lineplot

```python
sns.lineplot(data = dataset,
             x = "model_year",
             y = "mpg",
             hue = "origin")
```

Membuat scatterplot

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "mpg")
```
