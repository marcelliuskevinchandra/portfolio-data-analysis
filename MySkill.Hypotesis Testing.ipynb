```python
import pandas as pd
```

```python
diameter = pd.read_csv('diameter.csv')
diameter.head()
```

```python
H0 = 'pizza berdist normal'
H1 = 'pizza tidak berdist normal'
a = 0.05
```

```python
# Shapiro
from scipy.stats import shapiro
```

```python
shapiro(diameter) #p-value = 0.52 > 0.05 -> H0 tidak ditolak
```

```python
p_value = round(shapiro(diameter)[1],2)
p_value
```

```python
if p_value > a :
  print(H0)
else:
  print(H1)
```

Membandingkan dua kelompok

```python
pizzas = pd.read_csv('pizzas.csv')
pizzas.head()
```

Uji normalitas

```python
unit_1 = pizzas['Making Unit 1']
unit_2 = pizzas['Making Unit 2']
```

```python
H0 = 'pizza berdist normal'
H1 = 'pizza tidak berdist normal'
```

```python
shapiro(unit_1) #pvalue=0.32
```

```python
shapiro(unit_2) #pvalue=0.52
```

```python
#lengkapi dengan IF
p_value1 = round(shapiro(unit_1)[1],2)
p_value2 = round(shapiro(unit_2)[1],2)
```

```python
if p_value1 > a :
  print(H0)
else:
  print(H1)
```

```python
if p_value2 > a :
  print(H0)
else:
  print(H1)
```

Uji variansi

```python
from scipy.stats import levene
```

```python
H0 = 'Unit 1 dan unit 2 punya variansi sama'
H1 = 'Unit 1 dan unit 2 tidak punya variansi sama'
levene(unit_1, unit_2)
```

```python
#Lengkapi dengan IF
p_value = round(levene(unit_1, unit_2)[1],2)
if p_value > a :
  print(H0)
else:
  print(H1)
```

2 Sample t-test for Equal Variance

```python
from scipy.stats import ttest_ind
```

```python
H0 = 'Unit 1 dan 2 tidak berbeda secara signifikan'
H1 = 'Unit 1 dan 2 berbeda secara signifikan'
ttest_ind(unit_1,unit_2)
```

```python
#Lengkapi dengan IF
p_value = round(ttest_ind(unit_1, unit_2)[1],2)
if p_value > a :
  print(H0)
else:
  print(H1)
```
